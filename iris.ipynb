```python
%matplotlib inline
# %matplotlib notebook

import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_style('darkgrid')
sns.set_palette(sns.color_palette('bright'))
```

```python
col = ['sepal length', 'sepal width', 'petal length', 'petal width', 'species']
data = pd.read_csv('iris.data', names=col)
data.head()
```

```python
data.shape
```

```python
import pandas_profiling as pdprof

pdprof.ProfileReport(data)
```

```python
sns.pairplot(data, kind='scatter', diag_kind='kde', palette=sns.color_palette("bright"))
```

```python
sns.pairplot(data, hue='species', kind='scatter', diag_kind='kde', palette=sns.color_palette("bright"))
```

```python
sns.lmplot(x='petal length', y='petal width', hue='species', data=data)
```
